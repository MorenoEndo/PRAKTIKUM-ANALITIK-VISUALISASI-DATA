##**Latar Belakang**
Industri manhwa, atau komik asal Korea Selatan, telah mengalami perkembangan pesat dalam beberapa tahun terakhir. Dengan meningkatnya popularitas platform digital seperti Webtoon, banyak manhwa yang mendapatkan basis penggemar yang besar di seluruh dunia.
Pemahaman mengenai faktor-faktor yang berkontribusi terhadap kesuksesan sebuah manhwa menjadi penting bagi penerbit, kreator, maupun platform distribusi.

##**Tujuan Bisnis**
Tujuan utama dari analisis dataset ini adalah untuk memahami faktor-faktor yang mempengaruhi popularitas dan kesuksesan sebuah manhwa. Analisis ini dapat membantu berbagai pemangku kepentingan, seperti:

1. Penerbit dan Platform Digital → Menentukan strategi pemasaran yang lebih baik dengan memahami tren yang disukai pembaca.

2. Kreator Manhwa → Mengidentifikasi elemen cerita atau gaya seni yang lebih disukai agar dapat menciptakan karya yang menarik perhatian pembaca.

3. Pembaca dan Penggemar → Menemukan rekomendasi manhwa berdasarkan preferensi umum dan peringkat.


##**Permasalahan Bisnis**
Berdasarkan tujuan bisnis di atas, beberapa permasalahan yang perlu dianalisis adalah:

1. Identifikasi Tren Genre yang Populer

- Genre apa yang paling banyak diminati oleh pembaca?

- Genre mana yang memiliki skor tertinggi?

- Apakah genre tertentu memiliki kecenderungan untuk mendapatkan lebih banyak pembaca?


2. Hubungan antara Popularitas dan Skor

- Apakah manhwa dengan skor tinggi selalu memiliki popularitas tinggi?

- Apakah ada outlier, yaitu manhwa dengan skor tinggi tetapi tidak populer, atau sebaliknya?


3. Pengaruh Status Publikasi terhadap Minat Pembaca

- Apakah manhwa yang masih publishing lebih populer dibandingkan yang sudah finished?

- Apakah pembaca lebih tertarik untuk membaca manhwa yang sudah tamat atau yang masih berlanjut?


4. Korelasi antara Panjang Cerita dan Kesuksesan

- Apakah manhwa dengan lebih banyak chapter cenderung lebih populer?

- Berapa rata-rata jumlah bab yang dimiliki oleh manhwa dengan skor tertinggi?


5. Analisis Faktor yang Mempengaruhi Jumlah Favorit

- Faktor apa yang paling mempengaruhi jumlah favorit? (Genre, skor, popularitas, atau jumlah bab?)

- Apakah ada pola tertentu dalam preferensi pembaca terhadap manhwa yang dijadikan favorit?

##**Kriteria Keberhasilan**
Keberhasilan dari analisis ini akan ditentukan berdasarkan temuan yang dapat memberikan manfaat bagi industri manhwa. Beberapa indikator keberhasilan meliputi:

Dengan pemahaman yang mendalam dari analisis ini, industri manhwa dapat memanfaatkan data untuk mengambil keputusan yang lebih strategis dalam mengembangkan, menerbitkan, dan memasarkan manhwa.

##MENGECEK OUTLIERS

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

###MENGATASI DENGAN CARA IMPUTASI

Unnamed:0

```python
columns_to_impute = ["Unnamed: 0"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

score

```python
columns_to_impute = ["score"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

**Pola Distribusi Jumlah Volume**
- Jika histogram menunjukkan bentuk distribusi normal (bell curve), berarti sebagian besar manhwa memiliki jumlah volume yang berada di tengah-tengah rentang yang ada.
- Jika ada puncak di awal (skewed right), berarti sebagian besar manhwa memiliki volume yang lebih sedikit, dan hanya sedikit yang memiliki volume panjang.
- Jika ada puncak di akhir (skewed left), berarti kebanyakan manhwa memiliki volume panjang.

**Apakah Mayoritas Manhwa Memiliki Volume Pendek atau Panjang?**
- Jika bar tertinggi berada di sisi kiri, berarti kebanyakan manhwa memiliki volume yang lebih sedikit (misalnya, 1-10 volume).
- Jika bar lebih merata atau memiliki puncak di tengah, berarti ada distribusi yang lebih seimbang antara manhwa volume pendek dan panjang.

**Identifikasi Outlier (Volume Ekstrem)**
- Jika ada sedikit bar di sisi kanan (ekstrem), ini berarti hanya beberapa manhwa yang memiliki jumlah volume sangat besar.
- Bisa jadi ada manhwa populer dengan volume sangat panjang, yang bisa jadi rekomendasi untuk pembaca yang mencari seri panjang.

**Tren dalam Industri Manhwa**
- Jika sebagian besar manhwa memiliki jumlah volume yang sedikit, ini bisa menunjukkan tren bahwa pembaca lebih menyukai manhwa yang pendek.
- Jika ada banyak manhwa dengan volume panjang, ini bisa berarti seri panjang tetap diminati, atau industri cenderung menerbitkan cerita yang berlangsung lama.
